###payment_type == voucher

Ver se existem compras feitas só com voucher.

```python
#Ver se existem compras feitas só com voucher.

#Fazer uma lista com os order_id de cada payment_type
#Criar uma lista com order_id repetidos entre as listas
#Comparar o len das listas

#Criar set das order_id de cada payment_type
lista_compras_com_voucher = set(df_order_payments.query("payment_type == 'voucher'")['order_id'].sort_values().tolist())
lista_compras_com_debito = set(df_order_payments_order_id_repetido.query("payment_type == 'debit_card'")['order_id'].sort_values().tolist())
lista_compras_com_credito = set(df_order_payments_order_id_repetido.query("payment_type == 'credit_card'")['order_id'].sort_values().tolist())
lista_compras_com_boleto = set(df_order_payments_order_id_repetido.query("payment_type == 'boleto'")['order_id'].sort_values().tolist())

#Transformar set em lista.
lista_compras_com_voucher_sem_repetidos = list (lista_compras_com_voucher)
lista_compras_com_debito_sem_repetidos = list (lista_compras_com_debito)
lista_compras_com_credito_sem_repetidos = list (lista_compras_com_credito)
lista_compras_com_boleto_sem_repetidos = list (lista_compras_com_boleto)

#Comparar lista de repetidos com lista original
```

```python
len(lista_compras_com_voucher_sem_repetidos)
```

```python
#Verificar se tem order_id com payment_type voucher e payment_type debit_card

#Criar lista com repetidos
repetidos_voucher_debito = []

#Verificar

for i in lista_compras_com_voucher_sem_repetidos:
  for j in lista_compras_com_debito_sem_repetidos:
    if(i==j):
      repetidos_voucher_debito.append(i)
      break

len(repetidos_voucher_debito)

#Não tem
```

```python
#Verificar se tem order_id com payment_type voucher e payment_type credit_card

#Criar lista com repetidos
repetidos_voucher_credito = []

#Verificar

for i in lista_compras_com_voucher_sem_repetidos:
  for j in lista_compras_com_credito_sem_repetidos:
    if(i==j):
      repetidos_voucher_credito.append(i)
      break

len(repetidos_voucher_credito)

#2245 order_id com payment_type voucher e payment_type credit_card
```

```python
#Verificar se tem order_id com payment_type debit_card e payment_type credit_card

#Criar lista com repetidos
repetidos_debito_credito = []

#Verificar

for i in lista_compras_com_debito_sem_repetidos:
  for j in lista_compras_com_credito_sem_repetidos:
    if(i==j):
      repetidos_debito_credito.append(i)
      break

len(repetidos_debito_credito)

#1 order_id com payment_type debit_card e payment_type credit_card
```

```python
#Verificar se tem order_id com payment_type voucher e payment_type boleto

#Criar lista com repetidos
repetidos_voucher_boleto = []

#Verificar

for i in lista_compras_com_voucher_sem_repetidos:
  for j in lista_compras_com_boleto_sem_repetidos:
    if(i==j):
      repetidos_voucher_boleto.append(i)
      break

len(repetidos_voucher_boleto)

#Não tem
```

```python
#Ver len de lista_compras_com_voucher_sem_repetidos e len de repetidos_voucher_credito, que são as que tem voucher

print(len(lista_compras_com_voucher_sem_repetidos) - len(repetidos_voucher_credito))

#1621 compras só voucher.
#Vou ver elas
```

```python
#Ver as compras só com voucher
#juntar as 2 listas numa só

lista_compras_so_com_voucher = lista_compras_com_voucher_sem_repetidos.copy()
lista_compras_so_com_voucher = list(set(lista_compras_so_com_voucher) - set(repetidos_voucher_credito))
```

```python
#Ver as compras com voucher

#Vou usar loc. Para isso preciso passar os id_order par ao index. vou criar um novo DF.

#Copiar DF
df_order_payments_order_id_index = df_order_payments.copy()

#Mudar o indice
df_order_payments_order_id_index.set_index('order_id', inplace = True)

#df_order_payments_order_id_index

#Selecionar as linhas
df_order_payments_order_id_index = df_order_payments_order_id_index.loc[lista_compras_so_com_voucher].groupby('order_id').sum().sort_values('payment_value', ascending = False)
```
